# Financial Analysis

 import library

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import datetime 
%matplotlib inline
```

 List of stock

```python
my_stocks = ["tsla", "meta", "goog","MSFT", "NflX"]
```

```python
def stock_analysis(stock_ticker):
    stock = yf.Ticker(stock_ticker)
    df = stock.history(period="30d")
    print(f"Technical analysis for {stock_ticker}")
    plt.figure(figsize=(15,10))
    plt.plot(df.index,df.Open)
    plt.show()
    print(f"Volume analysis for {stock_ticker}")
    plt.figure(figsize=(15,10))
    plt.plot(df.index, df.Volume)
    plt.show()
    print(df)
    print()
    print(f"Balance sheet of {stock_ticker}")
    print(stock.balance_sheet)
    print()     
    print(f"Profit and loss panel of {stock_ticker}")
    print(stock.financials)
    print()
    print(f"Cash flow of {stock_ticker}",)
    print(stock.cashflow)
    print()
for stock in my_stocks:
    stock_analysis(stock.upper())
```
